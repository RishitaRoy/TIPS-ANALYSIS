```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df=pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
df
```

```python
print("All Males who tips")
male= df.loc[df["sex"] == "Male"]
male
```

```python
print("properties of all the males who tips")
male.describe().T.style.set_properties(**{"background-color": "lightblue","color":"black","border": "1.5px  solid black"})
```

```python
print("Females who tips")
female= df.loc[df["sex"] == "Female"]
female
```

```python
print("properties of all females who tips")
female.describe().T.style.set_properties(**{"background-color": "lightblue","color":"black","border": "1.5px  solid black"})
```

```python
print("A graph analysing males and females who tips with respect to total bill")
sns.scatterplot(x='total_bill',y='tip',data=df,hue='sex')
```

```python
print("Analysing the time with maximum number of males and females")
sns.countplot(x='time',data=df)
```

```python
print("Analysing the time most visited by both males and females")
sns.jointplot(x='sex',y='time',data=df)
```

```python
print("Analysing the day visited by most number of customes")
mostcust= sns.countplot(x='day', data=df)

```

```python
print("Analysing the number of males and females who visited most on saturdays")
sns.jointplot(x='day',y='sex',data=df)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```
